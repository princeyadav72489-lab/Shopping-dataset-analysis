```python
import pandas as pd
import os

csv_file = os.path.join(path, "Combined_dataset.csv")

df = pd.read_csv(csv_file)

df.head()
```

```python
print("Missing values before:")
print(df.isnull().sum())

df = df.fillna("Not Available")

print("\nDataset shape:")
print(df.shape)
```

```python
print("Shape:", df.shape)
print(df.columns.tolist())
print(df.dtypes)
```

```python
high_rated = df[df["rating"] > 4]

print(high_rated.shape)

high_rated.head()
```

```python
print(df.isnull().sum())
```

```python
df = df.fillna("Not Available")

print(df.shape)
```

```python
print("Shape:", df.shape)

print("\nColumns:")
print(df.columns.tolist())

print("\nData Types:")
print(df.dtypes)

print("\nFirst 5 Rows:")
display(df.head())

print("\nLast 5 Rows:")
display(df.tail())
```

```python
high_rated = df[df["rating"] > 4]

print("Number of products with rating > 4:")
print(high_rated.shape)

high_rated.head()
```

```python
selected = df[["title", "category", "rating", "final_price"]]

selected.head()
```

```python
print("Duplicates Before:", df.duplicated().sum())

df = df.drop_duplicates()

print("Duplicates After:", df.duplicated().sum())
```

```python
df["quantity"] = 1
```

```python
df["final_price"] = (
    df["final_price"]
    .astype(str)
    .str.replace("₹", "", regex=False)
    .str.replace(",", "", regex=False)
    .str.replace('"', "", regex=False)
)

df["final_price"] = pd.to_numeric(df["final_price"], errors="coerce")

print(df["final_price"].head())
```

```python
df["total_amount"] = df["final_price"] * df["quantity"]

df[["title", "final_price", "quantity", "total_amount"]].head()
```

```python
df.to_csv("cleaned_shopping_dataset.csv", index=False)

print("Cleaned dataset saved successfully!")
```
