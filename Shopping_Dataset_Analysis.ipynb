```python
import pandas as pd
import os

csv_file = os.path.join(path, "Combined_dataset.csv")
df = pd.read_csv(csv_file)

df.head()
```

```python
print("Shape:", df.shape)

print("\nColumns:")
print(df.columns.tolist())

print("\nData Types:")
print(df.dtypes)

print("\nFirst 5 Rows:")
display(df.head())

print("\nLast 5 Rows:")
display(df.tail())
```

```python
print(df.isnull().sum())
```

```python
df = df.dropna()

print("New Shape:", df.shape)
```

```python
high_rated = df[df["rating"] > 4]

display(high_rated.head())
```

```python
selected_columns = df[["title", "category", "rating", "final_price"]]

display(selected_columns.head())
```

```python
print("Duplicates Before:", df.duplicated().sum())

df = df.drop_duplicates()

print("Duplicates After:", df.duplicated().sum())
```

```python
df["quantity"] = 1
df["total_amount"] = df["final_price"] * df["quantity"]

display(df[["title", "final_price", "quantity", "total_amount"]].head())
```

```python
df.to_csv("cleaned_shopping_dataset.csv", index=False)

print("Cleaned dataset saved successfully!")
```
